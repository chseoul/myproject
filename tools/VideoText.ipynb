```python
# 유튜브 자막 추출 라이브러리 불러오기
from youtube_transcript_api import YouTubeTranscriptApi

# 유튜브 영상 ID (URL에서 v= 뒤에 있는 부분)
video_id = "YnaSl1eXLPI"

# 해당 영상에서 자막 추출 (언어는 한국어로 설정)
transcript = YouTubeTranscriptApi.get_transcript(video_id, languages=["ko"])

# 추출한 자막을 하나의 문자열로 합치기
# (자막은 기본적으로 여러 문장으로 분리되어 있으므로 하나로 합쳐야 프롬프트에 넣기 편함)
script = ""
for item in transcript:
    script = script + item["text"] + " "

# 결과 확인
print(script)
```

```python
# 유튜브 정보 추출 라이브러리 불러오기
import yt_dlp

# 유튜브 영상 URL (전체 링크)
video_url = "https://www.youtube.com/watch?v=YnaSl1eXLPI"

# yt-dlp 옵션 설정
ydl_opts = {
    'skip_download': True,  # 영상 파일 다운로드 없이 정보만 추출
    'quiet': False,          # 실행 중 불필요한 로그 숨기기
    'forcejson': True       # 추출 결과를 json 형식으로 가져오기
}

# yt-dlp를 사용해 유튜브 영상 정보 추출
with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    info = ydl.extract_info(video_url, download=False)

# 추출한 메타데이터 확인
print("제목:", info['title'])
print("설명:", info['description'])
print("썸네일:", info['thumbnail'])
print("업로드일:", info['upload_date'])
```
